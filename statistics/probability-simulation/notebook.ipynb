# Probability Simulation & Law of Large Numbers

This notebook demonstrates probability concepts using simulation:
- Coin Toss Probability
- Law of Large Numbers
- Central Limit Theorem


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

coin_tosses = np.random.choice([0,1], size=1000)
# 1 = Heads, 0 = Tails

probability_heads = np.mean(coin_tosses)

print("Estimated Probability of Heads:", probability_heads)
```

```python
cumulative_heads = np.cumsum(coin_tosses)
trials = np.arange(1, 1001)

running_probability = cumulative_heads / trials

plt.figure(figsize=(8,5))
plt.plot(trials, running_probability)
plt.axhline(y=0.5, color='r', linestyle='--')
plt.title("Law of Large Numbers - Coin Toss")
plt.xlabel("Number of Trials")
plt.ylabel("Running Probability of Heads")
plt.show()
```

```python
sample_means = []

for _ in range(1000):
    sample = np.random.choice([0,1], size=30)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(8,5))
sns.histplot(sample_means, bins=30, kde=True)
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

## Key Insights

- As number of trials increases, probability approaches 0.5.
- Running average stabilizes over time (Law of Large Numbers).
- Distribution of sample means approximates normal distribution (CLT).
