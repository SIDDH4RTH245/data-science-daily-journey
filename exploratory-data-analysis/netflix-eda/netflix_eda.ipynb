# Netflix Exploratory Data Analysis (EDA)

This notebook explores trends, distributions, and patterns in the cleaned Netflix dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("cleaned_netflix.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
df["date_added"].value_counts().sort_index().plot(kind="line")
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type")
plt.title("Movies vs TV Shows Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries by Content")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["rating"].value_counts().plot(kind="bar")
plt.xticks(rotation=45)
plt.title("Ratings Distribution")
plt.show()
```

## Key Insights

- Netflix content rapidly increased after 2015.
- Movies dominate compared to TV Shows.
- United States contributes most content.
- TV-MA is the most common rating.
